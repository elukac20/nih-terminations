Use FAIN to identify grants on website

```python
import requests
import time
from requests.adapters import HTTPAdapter
from urllib3.util.retry import Retry

session = requests.Session()

retries = Retry(
    total=5,
    backoff_factor=1,
    status_forcelist=[429, 500, 502, 503, 504],
    allowed_methods=["GET", "POST"]
)

session.mount("https://", HTTPAdapter(max_retries=retries))

headers = {
    "User-Agent": "Mozilla/5.0",
    "Accept": "application/json"
}

def search_fain(fain):
    time.sleep(1)

    payload = {
        "filters": {
            "award_type_codes": ["02", "03", "04", "05"],
            "award_ids": [fain]
        },
        "fields": [
            "Award ID",
            "Recipient Name",
            "Award Amount",
            "Description"
        ]
    }

    try:
        r = session.post(
            "https://api.usaspending.gov/api/v2/search/spending_by_award/",
            json=payload,
            headers=headers,
            timeout=30
        )
        r.raise_for_status()

        data = r.json()

        if not data.get("results"):
            print(f"No results for {fain}")
            return "error"

        internal_id = data["results"][0]["internal_id"]

        r = session.get(
            f"https://api.usaspending.gov/api/v2/awards/{internal_id}/",
            headers=headers,
            timeout=30
        )
        r.raise_for_status()

        award_data = r.json()

        return award_data["cfda_info"][0]["cfda_objectives"]

    except Exception as e:
        print(f"Error for {fain}: {e}")
        return "error"


print(search_fain("2301CAPREP"))
```

```python
df_hhs['description'] = df_hhs.apply(lambda i: search_fain(i['FAIN']), axis=1)
```
